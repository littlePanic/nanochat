```python
import torch
import pickle
import torch.nn.functional as F

from huggingface_hub import hf_hub_download
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
repo_id = "littlePanic99/nanochat"

base_filename = "base/d20/model_021400.pt"
mid_filename = "mid/d20/model_000809.pt"
sft_filename = "sft/d20/model_000700.pt"
tokenizer_filename = "tokenizer/latest/tokenizer.pkl"
```

```python
def get_model_values(filename, key="transformer.wte.weight"):
    ckpt = hf_hub_download(repo_id=repo_id, filename=filename)
    values = torch.load(ckpt, map_location="cpu")[key]
    return values
    
```

```python
base_emb = get_model_values(base_filename)
mid_emb = get_model_values(mid_filename)
sft_emb = get_model_values(sft_filename)
```

```python
base_emb.shape, mid_emb.shape, sft_emb.shape
```

```python
with open(hf_hub_download(repo_id=repo_id, filename=tokenizer_filename), "rb") as f:
    tokenizer = pickle.load(f)
    
```

# Difference Analysis

```python
def diff_size_plot(changed, title):
    plt.hist(changed, bins=100)
    plt.title(title)
    plt.xlabel("Cosine similarity")
    plt.ylabel("Token count")
    plt.show()

def diff_scatter_plot(emb_1, emb_2, per_token_cos, top_n=100):
    topk = torch.topk(1.0 - per_token_cos, k=top_n).indices
    X = torch.cat([emb_1[topk], emb_2[topk]], dim=0).float().cpu().numpy()
    X_pca = PCA(n_components=2).fit_transform(X)
    plt.scatter(X_pca[:100, 0], X_pca[:100, 1], s=20, label="emb1")
    plt.scatter(X_pca[100:, 0], X_pca[100:, 1], s=20, label="emb2")
    plt.legend()
    plt.title("Top 100 Most Changed Token Embeddings")
    plt.show()


def token_str(enc, tid: int):
    return enc.decode_bytes([tid]).decode("utf-8", errors="replace")


def summarise_differences(emb_1, emb_2):
    e1 = F.normalize(emb_1, dim=1)
    e2 = F.normalize(emb_2, dim=1)
    
    per_token_cos = (e1 * e2).sum(dim=1)
    
    most_changed_token = per_token_cos.min()
    average_changed_token = per_token_cos.min()
    least_changed_token = per_token_cos.max()
    
    print(f"Most changed token: {most_changed_token:0.3f}")
    print(f"Average changed token: {average_changed_token:0.3f}")
    print(f"Least changed token: {least_changed_token:0.3f}")

    diff_size_plot(per_token_cos.float().cpu().numpy(), "Tokens similarity")
    if most_changed_token < 0.95:
        diff_size_plot(per_token_cos[per_token_cos < 0.95].float().cpu().numpy(), "Tokens similarity < 0.95")
        diff_scatter_plot(e1, e2, per_token_cos)

    topk = torch.topk(1.0 - per_token_cos, k=100).indices
    topk_sorted = topk[
        torch.argsort(per_token_cos[topk])
    ]
    for i in topk_sorted:
        tid = i.item()
        print(f"{per_token_cos[tid].item():0.3f} {tid:<8} {token_str(tokenizer, tid):<15}")

```

```python
summarise_differences(base_emb, mid_emb)
```

```python
summarise_differences(mid_emb,sft_emb)
```
